```python
import cv2
import matplotlib.pyplot as plt
import time
```

이미지 크기 및 픽셀 정보 확인

```python
image = cv2.imread('토익사진.jpg')  # 넘파이 배열 객체

# 픽셀 수 및 이미지 크기 확인
print(f"shape: {image.shape}")
print(f"size: {image.size}")
print(f"type: {type(image)}\n")

# 이미지 numpy 객체의 특정 픽셀 가져오기
px = image[100, 100]

# B, G, R 순서로 출력된다
# GrayScale인 경우는 B,G,R로 구분되지 않음
print(px)
```

특정 범위의 픽셀 변경

```python
image = cv2.imread('토익사진.jpg')

# 반복문 사용
start_time = time.time()
for i in range(480, 570):
    for j in range(300, 700):
        image[i,j] = [255, 255, 255]
print(f"--- {time.time()-start_time} seconds ---")

# 넘파이 배열 사용 -> 훨씬 빠름
start_time = time.time()
image[480:570, 300:700] = [0, 0, 0]
print(f"--- {time.time()-start_time} seconds ---")

plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.show()
```

ROI (Region Of Interest 관심있는영역) 추출 및 복사

```python
image = cv2.imread('토익사진.jpg')

# numpy slicing
roi = image[480:570, 300:700]

# ROI 단위로 이미지 복사
image[0:90, 0:400] = roi

plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.show()
```

픽셀별로 색상 다루기

```python
image = cv2.imread('토익사진.jpg')
image[:400, :, 0] = 0  # B 색상 모두 0으로 바꾸기
image[400:800, :, 1] = 0  # G 색상 모두 0으로 바꾸기
image[800:, :, 2] = 0  # R 색상 모두 0으로 바꾸기

plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.show()
```

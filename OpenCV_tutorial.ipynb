### OpenCV의 기본적인 함수

- cv2.imread(file_name, flag): 이미지를 읽어 Numpy 객체로 만드는 함수
 
 flag: 이미지를 읽는 방법 설정

 - IMREAD_COLOR: 이미지를 Color로 읽고, 투명한 부분 무시
 - IMREAD_GRAYSCALE: 이미지를 Grayscale로 읽기
 - IMREAD_UNCHANGED: 이미지를 Color로 읽고, 투명한 부분도 읽기

- cv2.imshow(title, image): 특정한 이미지를 화면에 출력하는 함수
 
 코랩에서는 matplotlib 으로 출력해야 한다.
- cv2.cvtColor(image, flag): 이미지를 색상 형태를 변경하는 함수

```python
import cv2
import matplotlib.pyplot as plt

# OpenCV: [B, G, R]  /  Matplotlib: [R, G ,B] 로 다르다!!!
img_basic = cv2.imread('토익사진.jpg', cv2.IMREAD_COLOR)
plt.imshow(cv2.cvtColor(img_basic, cv2.COLOR_BGR2RGB))  # OpenCV와 matplotlib의 색상의 순서가 다르므로 변환해주어야 한다
plt.show()

img_basic = cv2.cvtColor(img_basic, cv2.COLOR_BGR2GRAY)
plt.imshow(cv2.cvtColor(img_basic, cv2.COLOR_GRAY2RGB))
plt.show()
```

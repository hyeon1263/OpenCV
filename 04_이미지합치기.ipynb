```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

- 이미지를 합치는 두 가지 방법 (이미지 픽셀 별로 더하기)

1. cv2.add(): Saturation 연산 :

 0보다 작으면 0, 255보다 크면 255로 표현

2. np.add(): Modulo 연산 :

 256은 0, 257은 1로 표현

```python
image_1 = cv2.imread('윤석열.jpg')
image_2 = cv2.imread('이재명.jpg')
```

```python
plt.imshow(cv2.cvtColor(image_1, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
image_2 = cv2.resize(image_2, None, fx=2.0, fy=2.0, interpolation=cv2.INTER_CUBIC) # 세로, 가로를 2배씩 늘어나도록 확대
```

```python
plt.imshow(cv2.cvtColor(image_2, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
height, width = image_1.shape[:2]

M = np.float32([[1, 0, 50], [0, 1, 0]])
dst = cv2.warpAffine(image_1, M, (width, height))
```

```python
# 이미지 합치기

result = cv2.add(image_2, dst)
plt.imshow(cv2.cvtColor(result, cv2.COLOR_BGR2RGB))
plt.show()

result = image_2 + dst
plt.imshow(cv2.cvtColor(result, cv2.COLOR_BGR2RGB))
plt.show()
```

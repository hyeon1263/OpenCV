```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

- 이미지를 합치는 두 가지 방법 (이미지 픽셀 별로 더하기)

1. cv2.add(): Saturation 연산 :

 0보다 작으면 0, 255보다 크면 255로 표현

2. np.add(): Modulo 연산 :

 256은 0, 257은 1로 표현

```python
image_1 = cv2.imread('benz.jpg')
image_2 = cv2.imread('background.jpg')
```

```python
image_1.shape, image_2.shape
```

```python
plt.imshow(cv2.cvtColor(image_1, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
plt.imshow(cv2.cvtColor(image_2, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
image_2 = cv2.resize(image_2, dsize=(680,424), interpolation=cv2.INTER_AREA)  # dsize=(너비, 높이) 순으로 크기 지정
```

```python
# 이미지 합치기

result = cv2.add(image_1, image_2)
plt.imshow(cv2.cvtColor(result, cv2.COLOR_BGR2RGB))
plt.show()

result = image_1 + image_2   # np.add(image_1, image_2) 와 같다
plt.imshow(cv2.cvtColor(result, cv2.COLOR_BGR2RGB))
plt.show()
```
